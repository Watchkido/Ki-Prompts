# Duplikate
Duplikate, also doppelt vorkommende Instanzen, müssen entfernt werden.

```python
import pandas as pd
import numpy as np
```

Nun erstelle ich mir einen eigenen DataFrame als Beispieldatensatz.

```python
patients = pd.DataFrame({'sex': ['male', 'male', 'female', 'male', 'female', 'unknown', 'unknown'],
                'age': [35, 46, 42, 15, 36, 42, 42],
                'BMI': [20, 20, 19.5, 19, 28, 21, 21],
                'sports activity': ['medium', 'medium', 'high', 'low', 'low', 'high', 'high']})
patients
```

```python
patients.duplicated()
```

```python
patients.drop_duplicates()
```

```python
patients.drop_duplicates(subset=['BMI'])
```

```python
patients
```

Natürlich kann man unter subset auch mehrere Spalten getrennt durch ein Komma angeben.

```python
patients.drop_duplicates(subset=['BMI', 'sports activity'])
```

```python
patients
```

Nun wurde die erste doppelte Instanz gelöscht. Möchte man aber die letzte behalten und dafür die erste Instanz löschen, geht man wie folgt vor:

```python
patients.drop_duplicates(['BMI'], keep='last')
```
