# Kovarianz und Korrelation

Lage- und Streuungsmaße erlauben eine Untersuchung einzelner Merkmale. Bei der Datenanalyse interessieren einem aber auch die Zusammenhänge zwischen Merkmalen eines Datensets. Maße, die einem darüber Informationen geben, sind die Kovarianz und die Korrelation.

```python
import pandas as pd
import numpy as np 
import seaborn as sns

iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris);
```

```python
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species');
```

cov() und corr() sind die Methoden, um Korrelation und Kovarianz von DataFrames zu berechnen.

```python
iris.cov ()
```

```python
iris.corr()
```

Um sich die Korrelation einer Spalte ausgeben zu lassen, kann man corrwith() verwenden.

```python
iris.corrwith(iris.sepal_width)
```

Zur Visualisierung der Korrelationskoeffizienten eignen sich Heatmaps.

```python
sns.heatmap(iris.corr(), annot=True, cmap='Blues');
```
