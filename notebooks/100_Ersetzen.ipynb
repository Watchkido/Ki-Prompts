# Ersetzen von fehlerhaften oder fehlenden Werten
Identifiziert man fehlende oder fehlerhafte Werte in einen Datenset, kann man diese entweder l√∂schen oder imputieren (ersetzen).

```python
import pandas as pd
```

```python
missings=['na','?','n/a','-']
penguins=pd.read_csv ('short_penguins.csv', na_values=missings)
penguins
```

```python
median=penguins['flipper_length_mm'].median()
penguins['flipper_length_mm'].fillna(median, inplace=True)
penguins
```

Oder mit Hilfe von forward fill (ffill) und backward fill (bfill) den vorderen oder nachfolgenden Wert zum Ersetzen nutzen.

```python
penguins.fillna(method='bfill')
```

Oder man ersetzt einen fehlenden Wert direkt durch einen angenommenen Wert.

```python
penguins.loc[14,'sex'] = 'female'
penguins
```

```python
penguins.replace({'torgersen': 'Torgersen'})
```

```python
penguins.replace({'torgersen': 'Torgersen', 'Dream': 'Dreams'})
```
