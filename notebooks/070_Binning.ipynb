## Binning
Unter Binning wird die Unterteilung eines Merkmals in einzelne Intervalle, auch Bins genannt, verstanden. 


```python
import pandas as pd
import numpy as np
```

```python
movies = pd.DataFrame({'Title of Movie': ['Twice Shy', 'Young Adults', 'Four Times', 'Colorado'],
    'Director of the Movie':['Adam Hersh', 'Alex Cladiff', 'Marc Spencer', 'Susan Sheffield'],
    'duration':[138, 111, 158, 130],
    'frequency': [2, 1, 0, 2],
    'rating':[4,7,5,1]})
    
movies
```

```python
bin_names_rating = ['low', 'medium', 'high']
bin_names_rating
```

```python
movies['rating_category'] = pd.cut(movies['rating'], 3, labels=bin_names_rating)
movies
```

```python
bin_names_duration = ['long', 'super long']
```

```python
bins_calculated = np.linspace(movies['duration'].min(), movies['duration'].max(), 3)
bins_calculated
```

```python
movies['duration_category'] = pd.cut(movies['duration'], bins_calculated, labels=bin_names_duration ,include_lowest=True)
movies
```
